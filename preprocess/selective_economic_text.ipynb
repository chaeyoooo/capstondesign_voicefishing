```python
# —— 0) Colab + Google Drive 마운트 ——
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# —— 1) 라이브러리 import & 경로 설정 ——
import os
import json

SRC_DIR     = '/content/drive/MyDrive/NIKL_DIALOGUE_2024_JSON'
DST_DIR     = '/content/drive/MyDrive/NIKL_DIALOGUE_2024_JSON_FILT'
TARGET_TOPIC = "경제/재테크/부동산/금융"

os.makedirs(DST_DIR, exist_ok=True)

# —— 2) JSON 필터링 & 저장 ——
for fname in sorted(os.listdir(SRC_DIR)):
    if not fname.lower().endswith('.json'):
        continue

    in_path = os.path.join(SRC_DIR, fname)
    with open(in_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # document 리스트 중 topic이 TARGET_TOPIC인 것만 추출
    filtered_docs = [
        doc
        for doc in data.get('document', [])
        if doc.get('metadata', {}).get('topic') == TARGET_TOPIC
    ]

    if filtered_docs:
        new_data = {
            "id":       data.get("id"),
            "metadata": data.get("metadata"),
            "document": filtered_docs
        }
        out_path = os.path.join(DST_DIR, fname)
        with open(out_path, 'w', encoding='utf-8') as f:
            json.dump(new_data, f, ensure_ascii=False, indent=2)
        print(f"✔ Saved filtered JSON: {fname} ({len(filtered_docs)} docs)")

print("모두 완료되었습니다.")
```
