```python
# ——— 0) 구글 드라이브 연동 ———
from google.colab import drive
drive.mount('/content/drive')

# ——— 1) whisper 설치 ———
!pip install openai-whisper

import whisper
import os
import pandas as pd

# ——— 2) 모델 로드 ———
model = whisper.load_model("small")

# ——— 3) unnormal 데이터 경로 ———
UNNORMAL_DIR = '/content/drive/MyDrive/unnormal/trimmed_samples'

# ——— 4) 파일 리스트 가져오기 ———
unnormal_files = [
    os.path.join(UNNORMAL_DIR, f)
    for f in os.listdir(UNNORMAL_DIR)
    if f.endswith('.wav')
]

# ——— 5) STT 추론 ———
stt_results = []

for i, file_path in enumerate(unnormal_files):
    result = model.transcribe(file_path, language='ko')
    text = result['text']
    stt_results.append({
        'file': os.path.basename(file_path),
        'transcribed_text': text
    })
    print(f"[{i+1}/{len(unnormal_files)}] {file_path} → {text}")

# ——— 6) 결과 저장 ———
df_stt = pd.DataFrame(stt_results)
df_stt.to_csv('/content/drive/MyDrive/unnormal_texts.csv', index=False)
```
